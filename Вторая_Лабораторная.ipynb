```
# подключение библиотек
import pandas as pd     
import numpy as np      
import matplotlib.pyplot as plot
import matplotlib.ticker as ticker 
```

```
# таблица значений силы тока
url = "https://drive.google.com/uc?export=download&id=10X7i5PYwSAe08JtXUKDBkfWUHTuI8NMm "

input_df = pd.read_csv(url, sep = ',',header = None)
input_df.head()
```

```
input_df.tail()
```

```
axes = input_df.plot(marker=None, figsize=(20, 20), subplots = False)
```

```
input_matrix = input_df.values # перевод таблицы в матрицу

print('размерность матрицы', input_matrix.shape)
print()

# среднее значение мощности, как полученное полученное по формуле P=R*I**2, где вместо I подставлено среднее значение модуля тока на рассматриваемом интервале

# цикл от 1 до числа столбцов
for i in range(0, len(input_df.columns)):
  print("{: <12} mean = {:.2f}".format(input_df.columns[i], 10*input_matrix[ 1280 , i].mean()**2))
```

```
# график значений мощности
fig, ax = plot.subplots(1, 1, figsize = (14, 8))

timestamps = range(0, input_matrix.shape[0], 5)

# зададим диапазон
begin = 0
end   = 1280

# легко указывать цвета, стиль линий, размеры и тип маркера
ax.plot(timestamps[begin : end], 10*input_matrix[begin : end, 0]**2, 'red', marker = '.', linewidth = 3, markersize = 9)

ax.set_ylabel('Мощность, Вт')
ax.legend(['Значение мощности'])

plot.grid(True)
plot.show()
```

```
# таблица значений мощности
url = "https://drive.google.com/uc?export=download&id=10X7i5PYwSAe08JtXUKDBkfWUHTuI8NMm "

input_df = 10*pd.read_csv(url, sep = ',',header = None)**2
input_df.head()
```

```
input_matrix = input_df.values # перевод таблицы в матрицу

print('размерность матрицы', input_matrix.shape)
print()

# среднее значение мощности как среднеарифметическое значений графика мощности

# цикл от 1 до числа столбцов
for i in range(0, len(input_df.columns)):
  print("{: <12} mean = {:.2f}".format(input_df.columns[i], input_matrix[ 1280 , i].mean()))
```
